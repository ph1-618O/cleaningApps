```python
#making a dictionary of month info and a search month function
feb = 28
jan = mar = may = jul = aug = octo = dec = 31
apr = jun = sep = nov = 30
#///////////////////////////////////////////////////////////////////////////////////////////////////////////////////
monthDict = {
    'january_01':jan,
    'february_02':feb,
    'march_03':mar,
    'april_04': apr,
    'may_05': may,
    'june_06': jun,
    'july_07':jul,
    'august_08':aug,
    'september_09': sep,
    'october_10':octo,
    'november_11': nov,
    'december_12':dec
                }
january = []
february = []
march= []
april=[]
may=[]
june=[]
july=[]
august=[]
september=[]
october=[]
november=[]
december=[]
months = [january, february, march, april, may, june, july, august, september, october, november, december]
#///////////////////////////////////////////////////////////////////////////////////////////////////////////////////
def searchDict(dictName, whatNum):
        for k, v in dictName.items():
            searchKey = k.split('_')
#             print(searchKey)
            if len(str(whatNum)) == 1:
                whatNum = '0'+ str(whatNum)
                if searchKey[1] == whatNum:
                    return v, searchKey[0], searchKey[1]
            elif searchKey[1] == str(whatNum):
                    return v, searchKey[0], searchKey[1]
#///////////////////////////////////////////////////////////////////////////////////////////////////////////////////
allIntervals = []
for i in range(len(months)+1):
    if (i == 0 & i < 12):
        i += 1
        intervals = searchDict(monthDict, i)
        monthKey = intervals[1]
        month = int(intervals[2])
        last_day_of_month = str(intervals[0])
        if len(str(month)) == 1: 
            startDay = str(f'2020-0{month}-01')
            endDay = str(f'2020-0{month}-{last_day_of_month}')
            print(f'1-{startDay}, {endDay}')
            allIntervals.append([startDay, endDay])
    if (i != 12 and i != 1):
        intervals = searchDict(monthDict, i)
        monthKey = intervals[1]
        month = int(intervals[2])
        last_day_of_month = str(intervals[0])
        if len(str(month)) == 1: 
            startDay = str(f'2020-0{month}-01')
            endDay = str(f'2020-0{month}-{last_day_of_month}')
            print(f'3-{startDay}, {endDay}')
            allIntervals.append([startDay, endDay])
        else:
            startDay = str(f'2020-{month}-01')
            endDay = str(f'2020-{month}-{last_day_of_month}')
            print(f'4-{startDay}, {endDay}')
            allIntervals.append([startDay, endDay])
    elif i == 1:
        continue
    else:
        i == 12
        intervals = searchDict(monthDict, i)
        monthKey = intervals[1]
        month = int(intervals[2])
        last_day_of_month = str(intervals[0])
        startDay = str(f'2020-{month}-01')
        endDay = str(f'2020-{month}-{last_day_of_month}')
        print(f'5-{startDay}, {endDay}')
        allIntervals.append([startDay, endDay])
```
